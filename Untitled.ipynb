```python
x = df.drop('exam_score',axis=1)
y = df['exam_score']

xtrain, xtest, ytrain, ytest = train_test_split(
    x, y, test_size=0.3, random_state=42
)
```

```python
num_cols = x.select_dtypes(include=np.number).columns
cat_cols = x.select_dtypes(include=['str','object']).columns

num_trans = Pipeline(steps =[
    ('scaler',RobustScaler(quantile_range=(20, 80)))
])
cat_trans = Pipeline(steps=[
    ('Label',OneHotEncoder(sparse_output=False, handle_unknown='ignore'))
])

preprocess = ColumnTransformer(transformers=[
    ('Num', num_trans, num_cols),
    ('Cat', cat_trans, cat_cols)
])
```

```python
models = {
    'Dummy': DummyRegressor(),
    'Linear': LinearRegression(),
    'RF': RandomForestRegressor(random_state=42),
    'XG': XGBRegressor(random_state=42),
    'LGBM': LGBMRegressor(random_state=42)
}

best_score = float('-inf')

best_model = None
best_model_name = ''

for name, model in models.items():

    pip = Pipeline(steps=[
        ('Preprocess', preprocess),
        ('model', model)
    ])

    print(f'============= {name} =============')

    scores = cross_val_score(
        pip,
        xtrain,
        ytrain,
        cv=5,
        scoring='r2'
    )
    avg_score = scores.mean()
    
    pip.fit(xtrain, ytrain)
    y_pred = pip.predict(xtest)

    res = ytest - y_pred
    r2 = r2_score(ytest, y_pred)
    mse = mean_squared_error(ytest, y_pred)
    mae = mean_absolute_error(ytest, y_pred)
    rmse = np.sqrt(mse)
    print(f'CV Scores: {scores}')
    print(f'Average CV Score: {avg_score:.4f}')
    print(f'R2 Score: {r2:.4f}')
    print(f'MSE: {mse:.4f}')
    print(f'RMSE: {rmse:.4f}')
    print(f'MAE: {mae:.4f}')
    res.hist()
    plt.title('Resisdual')
    plt.tight_layout()
    plt.show()
    if r2 > best_score:
        best_score = r2
        best_model_name = name
        best_model = pip

print(f'Best Model: {best_model_name}')
print(f'Best R2 Score: {best_score:.4f}')
joblib.dump(best_model, 'best_model.pkl')
```

```python
x_dl = data.drop('exam_score', axis=1)
y_dl = data['exam_score']

xtrain, xtest, ytrain, ytest = train_test_split(x_dl, y_dl,
                                               test_size=0.3, random_state=42)
```

```python
scaler = RobustScaler(quantile_range=(20, 80))
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
xtrain_t = torch.tensor(np.array(xtrain), dtype=torch.float32)
xtest_t = torch.tensor(np.array(xtest), dtype=torch.float32)
ytrain_t = torch.tensor(np.array(ytrain), dtype=torch.float32).unsqueeze(1)
ytest_t = torch.tensor(np.array(ytest), dtype=torch.float32).unsqueeze(1)
```

```python
n_features = xtrain.shape[1]
class student(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n_features, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(16, 1)
        )
    def forward(self, x):
        return self.net(x)

model = student(n_features)
total_para = sum(p.numel() for p in model.parameters())
print(model)
print(f'Total Parameters: {total_para}')
```

```python
criterion = nn.MSELoss()
optimizer = optim.AdamW(model.parameters(), lr=0.001)

class early_stopping:
    def __init__(self, patience = 25):
        self.patience = patience
        self.counter = 0
        self.best_weights = None
        self.best_loss = float('inf')

    def check(self, test_loss, model):
        if test_loss < self.best_loss:
            self.best_loss = test_loss
            self.counter = 0
            self.best_weights = {
                k: v.clone() for k, v in model.state_dict().items()
            }
            return False
        else:
            self.counter += 1
            return self.counter >= self.patience
```

```python
early_stop = early_stopping(patience=25)
test_losses = []
train_losses = []
stopped_epoch = 300

for epoch in range(1000):
    model.train()
    optimizer.zero_grad(set_to_none=True)
    train_pred = model(xtrain_t)
    train_loss = criterion(train_pred, ytrain_t)
    train_loss.backward()
    optimizer.step()
    train_losses.append(train_loss.item())

    model.eval()
    with torch.no_grad():
        test_pred = model(xtest_t)
        test_loss = criterion(test_pred, ytest_t)
        test_losses.append(test_loss.item())

    if epoch % 10 == 0:
        print(f'Epoch: {epoch} | Train_Loss: {train_loss.item():0.4f} | Test_Loss: {test_loss.item():0.4f}')

    if early_stop.check(test_loss.item(), model):
        stopped_epoch = epoch
        print(f'Stopped Epoch: {stopped_epoch:4d}')
        break

model.load_state_dict(early_stop.best_weights)
print(f'Best_Loss: {early_stop.best_loss}')
```

```python
model.eval()
with torch.no_grad():
    preds = model(xtest_t)
    preds = preds.numpy()

r2 = r2_score(ytest_t, preds)
mse = mean_squared_error(ytest_t, preds)
rmse = np.sqrt(mse)
mae = mean_absolute_error(ytest_t, preds)

print(f'R2 Score: {r2:.4f}')
print(f'MSE: {mse:.4f}')
print(f'RMSE: {rmse:.4f}')
print(f'MAE: {mae:.4f}')

plt.figure(figsize=(8, 4))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses,  label='Test Loss')
plt.axvline(stopped_epoch, color='red',
            linestyle='--', label=f'Early Stop @ {stopped_epoch}')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curve")
plt.legend()
plt.show()

torch.save({
    'model_state': model.state_dict(),
    'best_loss':   early_stop.best_loss,
    'n_features':  n_features,
}, 'dl_model.pth')

joblib.dump(scaler, 'scaler.pkl')
```
